```python
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
from faker import Faker
from PyPDF2 import PdfReader
import os

# Define page margins
LEFT_MARGIN = 50
RIGHT_MARGIN = 550
TOP_MARGIN = 750
BOTTOM_MARGIN = 50

def generate_pdf(filename, data):
    c = canvas.Canvas(filename, pagesize=letter)

    # Set font and size
    c.setFont("Helvetica", 12)

    # Split data into lines for word wrapping
    lines = data.split('\n')

    # Set starting y position
    y = TOP_MARGIN

    # Write data to PDF with word wrapping and page margins
    for line in lines:
        if line.strip():  # Check if line is not empty
            # Calculate text width
            text_width = c.stringWidth(line, "Helvetica", 12)

            # Check if the line fits within the right margin
            if text_width <= (RIGHT_MARGIN - LEFT_MARGIN):
                c.drawString(LEFT_MARGIN, y, line)
                y -= 15  # Adjust the line spacing if needed
            else:
                # Calculate number of characters that fit within the right margin
                max_chars = int(((RIGHT_MARGIN - LEFT_MARGIN) / text_width) * len(line))

                # Draw the line with adjusted width
                c.drawString(LEFT_MARGIN, y, line[:max_chars])
                y -= 15  # Adjust the line spacing if needed

                # Start a new line with the remaining text
                if max_chars < len(line):
                    remaining_text = line[max_chars:]
                    c.drawString(LEFT_MARGIN, y, remaining_text)
                    y -= 15  # Adjust the line spacing if needed

            # Check if new page is needed
            if y < BOTTOM_MARGIN:
                c.showPage()
                y = TOP_MARGIN

    c.save()
    
def fill_template(template, data):
    # Replace placeholders with data
    placeholders = ["[Date]", "[Producer/Content Creator Name]", "[Address]",
                    "[Media Company Name]", "[Description of the Content]",
                    "[Format of the Content]", "[Duration of the Content]",
                    "[Method of delivering the Content to Media Company]",
                    "[Amount]", "[Percentage]", "[State/Country]"]

    for placeholder, value in zip(placeholders, data):
        template = template.replace(placeholder, value)

    return template
def generate_fictional_data():
    faker = Faker()
    fictional_data = []

    for _ in range(num_documents):
        data = {
            "Date": str(faker.date_this_year()),
            "Producer/Content Creator Name": faker.name(),
            "Address": faker.address(),
            "Media Company Name": faker.company(),
            "Description of the Content": faker.sentence(),
            "Format of the Content": faker.word(),
            "Duration of the Content": str(faker.random_int(min=1, max=60)),  # Convert to string
            "Method of delivering the Content to Media Company": faker.word(),
            "Amount": str(faker.random_number(digits=5)),  # Convert to string
            "Percentage": f"{faker.random_int(min=5, max=25)}%",
            "State/Country": faker.state()
        }
        fictional_data.append(data)

    return fictional_data



def main():
    global num_documents
    num_documents = int(input("Enter the number of documents to be created: "))

    # Read template from file
    with open("template.txt", "r") as file:
        template = file.read()
       

    # Create a directory to store the generated PDFs
    if not os.path.exists("generated_pdfs"):
        os.makedirs("generated_pdfs")

    # Generate fictional data for each document
    fictional_data = generate_fictional_data()
    print(fictional_data)

    # Generate PDFs
    for i, data in enumerate(fictional_data, start=1):
        filled_template = fill_template(template, list(data.values()))
        filename = f"generated_pdfs/Media_Rights_Agreement_{i}.pdf"
        generate_pdf(filename, filled_template)

    print(f"{num_documents} PDFs generated successfully in the 'generated_pdfs' folder.")

if __name__ == "__main__":
    main()
```
